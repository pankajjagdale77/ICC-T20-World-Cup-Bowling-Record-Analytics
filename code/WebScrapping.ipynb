```python
#Importing Required Libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

### Data 2007

```python
web_data=requests.get("https://stats.espncricinfo.com/twenty20wc/engine/records/bowling/most_wickets_career.html?id=3115;type=tournament")
soup=BeautifulSoup(web_data.text,"html.parser")

Bowlers=soup.find("tbody").find_all("tr")
data_2007=[]
i=1
for Bowler in Bowlers:
    data=Bowler.select("td")
    
    Player_name=Bowler.find("td", class_="left").text.split("(")[0]
    Country_name=Bowler.find("td", class_="left").text.split("(")[1].rstrip(")")
    matches=Bowler.find("td", class_="padAst").text
    Innings=data[2].get_text()
    overs=data[3].get_text()
    mdn_overs=data[4].get_text()
    Runs=data[5].get_text()
    wkts=data[6].get_text()
    BBI=data[7].get_text()
    Ave=data[8].get_text()
    Econ=data[9].get_text()
    SR=data[10].get_text()
    four=data[11].get_text()
    five=data[12].get_text()
    
    data_2007.append({ "sr.no":i,
                       "Year":2007,
                       "Player_Name":Player_name,
                       "Country_Name":Country_name,
                       "Matches":matches,
                       "Innings":Innings,
                       "overs":overs,
                       "mdn_overs":mdn_overs,
                       "Runs":Runs,
                       "wkts":wkts,
                       "BBI":BBI,
                       "Ave":Ave,
                       "Econ":Econ,
                       "SR":SR,
                       "four":four,
                       "five":five })
    i=i+1
df_2007=pd.DataFrame(data_2007)
df_2007
```
